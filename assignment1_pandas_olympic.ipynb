__7. Is it true that Spain won fewer medals than Italy at the 2016 Olympics? Do not consider NaN values in _Medal_ column.__ 

*For discussions, please stick to [ODS Slack](https://opendatascience.slack.com/), channel #mlcourse_ai_news, pinned thread #a1_q6-10_fall2019*

- Yes
- No

```python
# You code here
# Правда ли, что Испания выиграла меньше медалей, чем Италия на Олимпиаде-2016? 
# Не учитывайте значения NaN в столбце медаль.
spainMedal = data[(~data['Medal'].isnull()) & (data['Year'] == 2016) & (data['Team'] == 'Spain')]
italyMedal = data[(~data['Medal'].isnull()) & (data['Year'] == 2016) & (data['Team'] == 'Italy')]
print('Spain', len(spainMedal), 'Italy', len(italyMedal))
```

__8. What age category did the fewest and the most participants of the 2008 Olympics belong to?__

*For discussions, please stick to [ODS Slack](https://opendatascience.slack.com/), channel #mlcourse_ai_news, pinned thread #a1_q6-10_fall2019*

- [45-55] and [25-35) correspondingly
- [45-55] and [15-25) correspondingly
- [35-45) and [25-35) correspondingly
- [45-55] and [35-45) correspondingly

```python
# You code here
# К какой возрастной категории относилось меньше всего 
# и больше всего участников Олимпиады 2008 года?
# 0-[15-25)
# 1-[25-35)
# 2-[35-45)
# 3-[45-55]
data['ageRange'] = None

def getAgeRange(x):
    if 15 <= x < 25:
        return 0
    elif 25 <= x < 35:
        return 1
    elif 35 <= x < 45:
        return 2
    elif 45 <= x <= 55:
        return 3
    else:
        return None
    

data['ageRange'] = data['Age'].apply(getAgeRange)
data.head()
```

```python
data[data['Year'] == 2008]['ageRange'].value_counts()
```
